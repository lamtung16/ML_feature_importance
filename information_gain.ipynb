```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import mutual_info_regression
import matplotlib.pyplot as plt
```

```python
# Create your dataset
np.random.seed(42)
x1 = np.random.rand(100)
x2 = np.random.rand(100)
noise = np.random.rand(100)
y = x1 + x2**2

# Create a DataFrame
data = pd.DataFrame({'x1': x1, 'x2': x2, 'x3': noise, 'y': y})
```

```python
# Extract features and target
X = data[['x1', 'x2', 'x3']]
y = data['y']

# Calculate information gain
information_gain = mutual_info_regression(X, y)
```

```python
# Print the information gain for each feature
for feature, gain in zip(X.columns, information_gain):
    print(f"{feature}: {gain:.4f}")
```

```python
plt.figure(figsize=(8, 6))
plt.bar(data.columns[:-1], information_gain)
plt.xlabel("Feature")
plt.ylabel("Importance")
plt.title("Information Gain")
plt.savefig("figs/information_gain.png", format='png')
plt.show()
```
